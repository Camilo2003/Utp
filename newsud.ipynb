#Inicializar tablero en vars


```python
# Define the domain of numbers (1-9)
dom = set(range(1, 10))

# Define the columns (A, B, C, ..., I) as a string
cols = "ABCDEFGHI"

def resetBoard(vars):
  for i in cols:
    for j in dom:
      vars[f'{i}{j}']=dom.copy()

def setValueInVars(vars,varName,varValue):
  vars[varName]={varValue}

def initBoard(vars,nameFile):
    file1 = open(nameFile, "r")
    ListSudoku = "ABCDEFGHI"
    for r in range(1,10):
      for c in ListSudoku:
        cur=int(file1.readline())
        if(cur<10):
          setValueInVars(vars,f"{c}{r}",cur)
```

#Creacion de restricciones

```python
# Define an empty dictionary to store constraints
const={'allDif':[]}

# Define a function to create column constraints
def colsConst(const):
  for i in cols:
    constVarList=[]
    for j in dom:
      constVarList.append(f'{i}{j}')
    const['allDif'].append(constVarList)


# Define a function to create row constraints
def rowsConst(const):
  for i in dom:
    constVarList=[]
    for j in cols:
      constVarList.append(f'{j}{i}')
    const['allDif'].append(constVarList)


# Define a function to create box (3x3) constraints
def winsConst(const):
    cols = ["ABC", "DEF", "GHI"]
    rows=["123","456","789"]
    for winRow in rows:
        for winCol in cols:
            window = []
            for row in winRow:
                for col in winCol:
                    window.append(f"{col}{row}")
            const["allDif"].append(window)

# Call constraint functions to create constraints
colsConst(const)
rowsConst(const)
winsConst(const)
const
```

##Alldif

```python
# Define a function to enforce the "allDif" constraint
"""
    Enforce the "allDif" constraint to eliminate possibilities in cells.

    Parameters:
        vars (dict): A dictionary representing the Sudoku board.
        varsList (list): A list of variables to apply the constraint.
        verbose (bool): Enable verbose mode for debugging (default: False).

    Returns:
        bool: True if any changes were made, False otherwise.
"""
def allDif(vars,varsList,verbose=False):
  anyChange=False
  for var in varsList:
    if(len(vars[var])==1):
      for j in  varsList:
        if not(j==var) and len(vars[j])>1:
          oldLen=len(vars[j])
          oldDom=vars[j].copy()
          vars[j].discard(list(vars[var])[0])
          newDom=vars[j].copy()
          newLen=len(vars[j])
          if not(oldLen==newLen):
            anyChange = True
            if verbose:
              print(f'allDif in var({j})\toldDom({oldDom})\t\t\tnewDom({newDom})\t\t\tRemoveFromVar({var})\tdomVar({vars[var]})')
  return anyChange
```

##exval

```python
# Define a function to enforce the "exval" constraint
"""
    Enforce the "exval" constraint to assign values to cells with unique possibilities.

    Parameters:
        vars (dict): A dictionary representing the Sudoku board.
        varsList (list): A list of variables to apply the constraint.
        verbose (bool): Enable verbose mode for debugging (default: False).

    Returns:
        bool: True if any changes were made, False otherwise.
"""
def exval(vars,varslist,verbose=False):
  anychange=False
  for var in varslist:
    setDom=set()
    othersDom=set()
    if(len(vars[var])>1):
      setDom=vars[var].copy()
      for j in varslist:
        if not(j==var):
          othersDom=othersDom | vars[j].copy()
      oneDom=setDom-othersDom
      if(len(oneDom)==1):
        oldLen=len(vars[var])
        oldDom=vars[var].copy()
        vars[var]=oneDom.copy()
        newLen=len(vars[var])
        newDom=vars[var].copy()
        anychange=True
        if verbose:
          print(f'exval in var({var})\toldDom({oldDom})\t\t\tnewDom({newDom}')
  return anychange
```

##TwoDom

```python
# Define a function to enforce the "twoDom" constraint
"""
    Enforce the "twoDom" constraint to eliminate possibilities in cells.

    Parameters:
        vars (dict): A dictionary representing the Sudoku board.
        varsList (list): A list of variables to apply the constraint.
        verbose (bool): Enable verbose mode for debugging.

    Returns:
        bool: True if any changes were made, False otherwise.
"""
def twoDom(vars, varsList,verbose):
  anyChange=False
  for var in varsList:
    dom_var=set()
    count_same_dom = 0
    if len(vars[var]) == 2:
        dom_var = vars[var]
    for var2 in varsList:
      if vars[var2] == dom_var:
        count_same_dom += 1
    if count_same_dom == 2:
      for j in varsList:
        if vars[var]!=vars[j]:
          oldLen=len(vars[j])
          oldDom=vars[j].copy()
          vars[j] = vars[j] - dom_var
          newDom=vars[j].copy()
          newLen=len(vars[j])
          if not(oldLen==newLen):
            anyChange = True
            if verbose:
              print(f'TwoDom in var({j})\toldDom({oldDom})\t\t\tnewDom({newDom})\t\t\tRemoveFromVar({var})\tdomVar({vars[var]})')
  return anyChange
```

#Consistencia global

```python
# Define a function to enforce global consistency using various constraint functions
"""
    Enforce global consistency using multiple constraint functions.

    Parameters:
        vars (dict): A dictionary representing the Sudoku board.
        const (dict): A dictionary containing constraints.
        funcs (list): A list of constraint functions to apply.
        verbose (bool): Enable verbose mode for debugging.

    Returns:
        None
"""
def globalConsist(vars,const,funcs,verbose):
  anychange=True
  while anychange:
    anychange=False
    for i in const['allDif']:
      for func in funcs:
        anychange=True if eval(f'{func}(vars,i,verbose)') else anychange
```

```python
# Define a function to find the empty cell with the minimum domain size
"""
    Find the empty cell with the minimum domain size.

    Parameters:
        vars (dict): A dictionary representing the Sudoku board.

    Returns:
        str: The name of the empty cell with the smallest domain size.
"""
def find_empty_cell_min_dom(vars):
  var_min_tamano=None
  tamano_min=10
  for var,dom in vars.items():
    tamano_dom=len(dom)
    if tamano_dom<tamano_min and tamano_dom>1:
      tamano_min=tamano_dom
      var_min_tamano=var
  return var_min_tamano
```

```python
# Define a function to check if a value is valid in a list of variables
"""
    Check if a value is valid in a list of variables.

    Parameters:
        varslist (list): A list of variables to check.
        var (str): The variable to validate.
        num (int): The value to validate.

    Returns:
        bool: True if the value is valid in the list of variables, False otherwise.
"""

def is_valid_in_list(varslist,var,num):
  dom={num}
  if var in varslist:
    for i in varslist:
      if not(var==i):
        if vars[i]==dom:
          return False
  return True
```

```python
# Define a function to check if a value is valid in all constraints
"""
    Check if a value is valid in all constraints.

    Parameters:
        const (dict): A dictionary containing constraints.
        var (str): The variable to validate.
        num (int): The value to validate.

    Returns:
        bool: True if the value is valid in all constraints, False otherwise.
"""
def is_valid(const,var,num):
  for varslist in const['allDif']:
    if not(is_valid_in_list(varslist,var,num)):
      return False
  return True
```

```python
# Define a function to solve the Sudoku using backtracking and constraints
"""
    Solve the Sudoku using backtracking and constraint propagation.

    Parameters:
        vars (dict): A dictionary representing the Sudoku board.
        const (dict): A dictionary containing constraints.
        funcs (list): A list of constraint functions to apply.
        verbose (bool): Enable verbose mode for debugging.

    Returns:
        bool: True if a valid solution is found, False otherwise.
"""
def solve_sudoku(vars,const,funcs,verbose):
    globalConsist(vars,const,funcs,verbose)
    # Encuentra una celda vacía
    empty_var = find_empty_cell_min_dom(vars)
    print(empty_var)

    # Si no hay celdas vacías, el Sudoku está resuelto
    if not empty_var:
        return True

    dom_var=vars[empty_var].copy()
    print(dom_var)

    for num in dom_var:
        print("ONE")
        if is_valid(const, empty_var, num):
            # Asigna el número si es válido
            vars[empty_var] = {num}

            # Intenta resolver el Sudoku con la asignación actual
            if solve_sudoku(vars,const,funcs,verbose):
                return True

            # Si la asignación no conduce a una solución válida, retrocede
            vars[empty_var] = dom_var

    # Si ningún número es válido en esta celda, retrocede
    return False
```
